```python
import numpy as np
import matplotlib.pyplot as plt

def uni(a,b,n):
  return np.random.uniform(a,b,n)

def pred(x):
  a_pred = a_pred=(np.mean(x)-np.sqrt(3*(np.mean(x**2)-(np.mean(x))**2)))
  b_pred = b_pred=(np.mean(x)+np.sqrt(3*(np.mean(x**2)-(np.mean(x))**2)))
  return a_pred,b_pred
```

```python
#задаем истинные значения и размери сэмпла
a = 0
b = 1
n = 500
x=uni(a,b,500) #
a_pred=pred(x)[0]
b_pred=pred(x)[1]

a_error=(a-a_pred)**2
b_error=(b-b_pred)**2

print(a_error)
print(b_error)
```

```python
a_errors = []
b_errors = []

for i in range(1,300):
  x=uni(a,b,i)
  a_pred=pred(x)[0]
  b_pred=pred(x)[1]
  a_errors.append((a-a_pred)**2)
  b_errors.append((b-b_pred)**2)
  
```

```python
plt.plot(a_errors,label='aerror',color='r')
plt.legend('aerror')
plt.show()
```

```python
plt.plot(b_errors,label='berror',color='r')
plt.legend('berror')
plt.show()
```
