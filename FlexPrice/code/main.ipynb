```python
import pandas as pd
```

```python
olist_customers_dataset = pd.read_csv('C:\\Users\\Admin\\Desktop\\AIM\\aim_2_hw\\FlexPrice\\data\\olist_customers_dataset.csv')
olist_geolocation_dataset = pd.read_csv('C:\\Users\\Admin\\Desktop\\AIM\\aim_2_hw\\FlexPrice\\data\\olist_geolocation_dataset.csv')
olist_order_items_dataset = pd.read_csv('C:\\Users\\Admin\\Desktop\\AIM\\aim_2_hw\\FlexPrice\\data\\olist_order_items_dataset.csv')

olist_order_payments_dataset = pd.read_csv('C:\\Users\\Admin\\Desktop\\AIM\\aim_2_hw\\FlexPrice\\data\\olist_order_payments_dataset.csv')
olist_order_reviews_dataset = pd.read_csv('C:\\Users\\Admin\\Desktop\\AIM\\aim_2_hw\\FlexPrice\\data\\olist_order_reviews_dataset.csv')
olist_orders_dataset = pd.read_csv('C:\\Users\\Admin\\Desktop\\AIM\\aim_2_hw\\FlexPrice\\data\\olist_orders_dataset.csv')

olist_products_dataset = pd.read_csv('C:\\Users\\Admin\\Desktop\\AIM\\aim_2_hw\\FlexPrice\\data\\olist_products_dataset.csv')
olist_sellers_dataset = pd.read_csv('C:\\Users\\Admin\\Desktop\\AIM\\aim_2_hw\\FlexPrice\\data\\olist_sellers_dataset.csv')
product_category_name_translation = pd.read_csv('C:\\Users\\Admin\\Desktop\\AIM\\aim_2_hw\\FlexPrice\\data\\product_category_name_translation.csv')
```

```python
olist_products_dataset = olist_products_dataset.merge(
    product_category_name_translation,
    on="product_category_name",
    how="left"
)

olist_products_dataset
```

```python
df = olist_order_items_dataset.copy()

# Добавляем заказы с датой покупки и customer_id
df = df.merge(
    olist_orders_dataset,
    on="order_id",
    how="left"
)
```

```python
# Добавляем характеристики товара
df = df.merge(
    olist_products_dataset,
    on="product_id",
    how="left"
)
```

```python
# Добавляем информацию о покупателях
df = df.merge(
    olist_customers_dataset,
    on="customer_id",
    how="left"
)
```

```python
# Добавляем продавцов
df = df.merge(
    olist_sellers_dataset,
    on="seller_id",
    how="left"
)
```

```python
# Добавляем платежи
df = df.merge(
    olist_order_payments_dataset,
    on="order_id",
    how="left"
)
```

```python
# Добавляем отзывы (оценки)
df = df.merge(
    olist_order_reviews_dataset,
    on="order_id",
    how="left"
)
```

```python
# Добавляем геолокацию покупателей (по ZIP)
df = df.merge(
    olist_geolocation_dataset,
    left_on='customer_zip_code_prefix',
    right_on='geolocation_zip_code_prefix',
    how='left'
)
```
