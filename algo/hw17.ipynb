```python
class Box:
    def __init__(self, value = None):
        self._value = value
        self._next_box = None

class LinkedList:
    def __init__(self):
        self._head = None

    def add(self, new_value):
        new_box = Box(new_value)

        if self._head is None:
            self._head = new_box
            return

        last_box = self._head

        while (last_box._next_box):
            last_box = last_box._next_box

        last_box._next_box = new_box

    def contains(self, value):
        last_value = self._head

        while(last_value):
            if last_value._value == value:
                return True
            last_value = last_value._next_box

        return False
    
    def get(self, index):
        last_box = self._head
        box_index = 0

        while (last_box):
            if box_index == index:
                return last_box._value
    
            last_box = last_box._next_box
            box_index += 1  

        raise ValueError("Index greater then lenght of list")
    
    def remove_by_value(self, rmvalue):
        cur = self._head
        prev = None

        while (cur):
            if cur._value == rmvalue:
                if prev:
                    prev._next_box = cur._next_box
                else:
                    self._head = cur._next_box
                return
            prev = cur
            cur = cur._next_box
            
        raise ValueError("Value not in list")
    
    def remove_by_index(self, rmindex):
        cur = self._head
        prev = None
        box_index = 0

        while (cur):
            if box_index == rmindex:
                if prev:
                    prev._next_box = cur._next_box
                else:
                    self._head = cur._next_box
                return 
            
            box_index += 1
            prev = cur
            cur = cur._next_box



        raise ValueError("Index greater then len of list")
        
```

```python
l = LinkedList()
l.add(1)
l.add(2)
l.add(3)
l.add(4)
l.add(5)
l.add(6)
l.add(7)
l.add(8)

print(l.contains(4))
l.remove_by_index(10)
print(l.contains(4))
```
