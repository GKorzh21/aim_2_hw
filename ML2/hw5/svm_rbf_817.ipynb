```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score

tr_sample = tr.sample(10000, random_state=42)

X_train, X_val, y_train, y_val = train_test_split(
    tr_sample[['model_1', 'model_2', 'model_3', 'model_4', 'model_5']],
    tr_sample['target'],
    test_size=0.2,
    random_state=42,
    stratify=tr_sample['target']
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)

svm_model = SVC(
    kernel='rbf',
    C=1.0,
    gamma='scale',
    class_weight='balanced',
    probability=True,
    random_state=42
)

print('Start training...')
svm_model.fit(X_train_scaled, y_train)
print('Training completed!')

print('Predicting on validation set...')
y_val_pred = svm_model.predict(X_val_scaled)
y_val_pred_proba = svm_model.predict_proba(X_val_scaled)[:, 1]

print("\nValidation Metrics:")
print(classification_report(y_val, y_val_pred))
print("ROC-AUC:", roc_auc_score(y_val, y_val_pred_proba))

print("\nPredicted class distribution (Validation):")
print(pd.Series(y_val_pred).value_counts(normalize=True))
```

```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, roc_auc_score

X_train = tr[['model_1', 'model_2', 'model_3', 'model_4', 'model_5']]
y_train = tr['target']

X_test = tst[['model_1', 'model_2', 'model_3', 'model_4', 'model_5']]

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


svm_model = SVC(
    kernel='rbf',          # Гауссово (RBF) ядро
    C=1.0,                # Параметр регуляризации
    gamma='scale',        # Масштабирование коэффициента ядра
    class_weight='balanced',  # Автоматическая балансировка классов
    probability=True      # Для получения вероятностей (если нужно)
)


print('start fit')
svm_model.fit(X_train_scaled, y_train)
print('end fit')


print('start pred')
y_pred = svm_model.predict(X_test_scaled)
y_pred_proba = svm_model.predict_proba(X_test_scaled)[:, 1]  # Вероятности класса 1
print('end pred')


# Вывод результатов (если есть true labels для теста)
# print(classification_report(y_test, y_pred))
# print("ROC-AUC:", roc_auc_score(y_test, y_pred_proba))

predictions = pd.DataFrame({
    'order_id': tst['order_id'],
    'predicted_class': y_pred,
    'predicted_probability': y_pred_proba
})
```

```python
t = 32
predictions[f'predicted_{t}'] = (y_pred_proba >= t / 100).astype(int)
predictions[predictions['predicted_class'] != predictions[f'predicted_{t}']]
```
