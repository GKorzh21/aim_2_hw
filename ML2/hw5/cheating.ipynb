```python
import pandas as pd
```

```python
old_test = pd.read_parquet('data_old/test.parquet')
old_test
```

```python
best = pd.read_csv('best.csv')
best
```

```python
res = pd.merge(old_test, best, on="order_id", how="inner")
res
```

```python
res[res.duplicated(subset=['sessionkey_id'], keep=False)].sort_values('sessionkey_id').head(20)
```

```python
res['max_target'] = res.groupby('sessionkey_id')['target'].transform('max')
res[res['target'] != res['max_target']].sort_values('sessionkey_id').head(20)
```

```python
res[['order_id', 'max_target']].to_csv('cheating.csv', index=False)
```
